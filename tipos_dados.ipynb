```python
#tipos de dados no python - revisão

#int
idade = 19
print("idade: ")
print(idade)
print(type(idade))

#float(ponto flutuante) valores com virgula (ponto separador de decimais)
altura = 1.67
print("altura: ")
print(altura)
print(type(altura))


#str (string) texto
nome = "Peter"
print("nome: ")
print(nome)
print(type(nome))


#bool - booleano: o que é um boleano galerinha (sim, não, verdade, mentira, ligado, desligado) ou seja só há duas opções

briga = False
print("briga: ")
print(briga)
print(type(briga))


#list Lista (coleção ordenada)
frutas = ["blueberry", "morango", "manga"]
print("frutas: ")
print(frutas)
print(type(frutas))



#tuple - dupla, como por exemplo, coordenadas
coordenadas = (52.640, 24.325)
print("coordenadas: ")
print(coordenadas)
print(type(coordenadas))


#set - conjunto
numeros = {1, 2, 2, 3, 3, 4, 4, 5, 6, 6}
idade = 19
print("numeros: ")
print(numeros)
print(type(numeros))


#dict - dicionario
pessoas = {"nome": "Hannie", "idade": 16}
print("pessoas: ")
print(pessoas)
print(type(pessoas))


#None Type = quando não tem nada, mas preciso informar um valor
valor = None
print("valor: ")
print(valor)
print(type(valor))
```
